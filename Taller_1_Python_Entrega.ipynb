## Pregunta 3

Defina una función **recursiva** que invierta textos. Es decir:
- Si le entra como argumento un `string` que lo invierta. 
``` python 
'Hola' -> 'aloH'
```
- Si le entra como argumento una lista de `strings` invierta cada `string` en la lista.
``` python 
['Hola', 'chao'] -> ['aloH', 'oahc']
```
- Si le entra como argumento una lista de listas de `strings` invierta cada `string` en cada una de las listas.
``` python 
['Hola', ['chao', 'adios']] -> ['aloH', ['oahc', 'soida']]
```

Llame a esta función `invertirRecursivo`. 

````python
m="Hola"
n=["Hola", "chao"]
l=["Hola", ["chao", "adios"]]
````

````python
def invertirRecursivo(lista):
    resultado2=[]     #Primero creamos una lista vacia donde se van a añadir los elementos
    if type(lista)==str:  #Entonces si el tipo de elemento de la lista es un string, se realiza el procedimiento usual de invertir
        return lista[::-1]  #solo se invierte la palabra
    elif type(lista)==list: #pero si el elemento ingresado es un list
        for i in lista:   #entonces se realiza un loop que analiza el tipo de elemento de cada lista y si es un string solo se invierte
            if type(i)==str:
                resultado2.append(i[::-1])  #aqui se encuentra la lista donde se adhieren las palabras ya invertidas
            elif type(i)==list: #pero si es una lista que esta dentro de la lista, se invierte las palabras dentro de la lista
                resultado3=[j[::-1] for j in i]
                resultado2.append(resultado3) #aqui hay una lista a la que se adhieren estas palabras
            
    print(resultado2) #Por último se imprime el resultado final
````

````python
invertirRecursivo(m)
````

````python
invertirRecursivo(n)
````

````python
invertirRecursivo(l)
````

```python
In: listaPrimos(10, imprimir=False)
Out: [2, 3, 5, 7]
```

```python

In: listaPrimos(10, imprimir=True)

El número 10 tiene 4 divisores

Out : [2, 3, 5, 7]
```



````python
#### RESPUESTA TOTAL #############################################

def listaPrimos(m, imprimir):
    if imprimir==True:                    #Para este primer apartado, si imprimir==True, solo si esto se cumple se realizará este
        listanumeros=list(range(2,m+1))   #este proceso. Lo que se realiza es crear una lista de numeros entre [2,3,4,..,m] con
        for j in listanumeros:            #la herramiento range(2,m+1); de ahi se crea el primer loop para esta lista, despues se
            listaDivisores=[1,j]          #abre otro loop para los divisores y despues se pone una condicion de que el residuo 
            for s in range(2,j):          #de la division de los elementos de la primera lista, con la segunda tienen que ser cero,
                if(j%s==0):               # si es así, se añade a la listaDivisores.Por último se imprime la oración requerida.
                    listaDivisores += [s]
            
            print("El número", j, "tiene", len(listaDivisores), "divisores")
        
    if imprimir==False or True:          #Este proceso se realizará si el valor de imprimir es igual a False o a True.Primero se crea
        pr=[]                            #una lista vacia pr=[], para ahí añadir los primos. y tmbn se incluye un contador que estará
        contador = 0                     #perimitiendo el seguimiento de los números. Con la ayuda de dos loops, uno incluyendo la lista
        for i in range(2, m+1):          #de números entre 2 y m, y la otra la lista de numeros  entre 2 y 11, esto ultimo serán los elementos
            primos=True                  #con los que se va a dividir los elementos de la primera lista.
            for l in range(2,11):        #Se utiliza la ayuda de primos=True y primos=False, para excluir aquellos elementos que
                if i == l:               #que son iguales o que no son primos.
                    break                #de ahi con el append.(i) se adhieren los elementos a la lista de pr, de primos.
                elif i%l == 0:
                    primos = False
                else:
                    continue
            if primos == True:
                pr.append(i)
            contador = contador + 1
        print(pr)
                
````

````python
# RESPUESTA
def mezclarPalabras(m,n):
    """
    Esta función se encarga de mezclar intercaladamente las letras de las dos palabras que puedes ser ingresadas en la función.
    
    Parametros:
    ----------
    
    m: str      n:str
    
        Son las palabras que se quieren mezclar intercaladamente. 
    
    """
    lista1=[i for i in m]  # Primero, lo que se quiso hacer es que con la lista 1 y 2 es separar cada palabra(string) en una lista 
    lista2=[j for j in n]  # por ejemplo: 'Hola' a ['H','o','l','a']. 
    lista3=lista1[::2]     # Con la creación de las listas 3 y 5 se sustrae aquellos elementos, o mejor dicho, letras que estan 
    lista4=lista1[1::2]    # en una posicion par 0,2,4, por eso el salto en dos empezando desde cero. [::2]
    lista5=lista2[::2]     # Con la creación de las listas 4 y 6, se realiza el mismo procedimiento, pero con las posiciones impares
    lista6=lista2[1::2]    
    listaprimera=[]
    i=0
    j=0
    while i<len(lista3)+1 and j<len(lista6):  #se crea un bucle, en el cual se adjunten las letras intercaladamente entre una lista a otra
        listaprimera.append(lista3[i]+lista6[j])
        i=i+1
        j=j+1
        
    palabra1=''.join(listaprimera)   #juntamos los valores de la lista para formar la primera palabra
    
    z=0
    x=0
    listasegunda=[]
    while z<len(lista5) and x<len(lista4):  #mismo procedimiento de creación de bucle
        listasegunda.append(lista5[z]+lista4[x])
        z=z+1
        x=x+1
        
    palabra2=''.join(listasegunda)  #se forma la segunda palabra
    
    listaterminada=[palabra1,palabra2]  #se adhiere ambas palabras a una lista, llamada: listaterminada y finalmente se imprime.
    
    print(listaterminada)
        
````

````python
mezclarPalabras(texto_1,texto_2)
````
