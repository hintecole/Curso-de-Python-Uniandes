```python
sumadead=DeadGlobal.drop(columns=['Province/State', 'Lat', 'Long'])
```

### Pregunta 2

```python
# Para esta respuesta se utiliza el método recomendado por dos artículos de revistas de medicina internacionales y
# y por un centro médico en Nebraska. Se calculará lo conocido como el case fatality rate para ver la mortalidad. 
#Artículo del Journal of Medical Systems: https://link.springer.com/article/10.1007/s10916-020-01668-6
#Centro Nebraska Medicine: https://www.nebraskamed.com/COVID/how-to-calculate-covid-19-stats-for-your-area#case_fatality_rate
#Artículo del Central Asian Journal of Global Health: http://cajgh.pitt.edu/ojs/cajgh/article/view/489/315
#
# La formula es (M/C)*100, donde M= fallecidos covid y C=numero de casos confirmados. 
```

```python
totalcasos=CasosGlobal.drop(columns=['Province/State', 'Lat', 'Long'])
```

```python
totalcasos['totalpositivos']=totalcasos.sum(axis=1)
```

```python
positivos=totalcasos.groupby('Country/Region')[['totalpositivos']].sum()
```

```python
positivos.reset_index(inplace=True)
fallecidos.reset_index(inplace=True)
```

```python
defo=pd.merge(fallecidos, positivos)
```

```python
defo
```

```python
defo['tasamortalidad']=100*defo['totalmuertos']/defo['totalpositivos']
```

```python
defo
```

```python
defo.tasamortalidad.sort_values()
```

```python
defo.iloc[189]
#El país con la mayor tasa de mortalidad (case fatality rate) es Yemen.
```

### Pregunta 3

```python
# Para ver la tasa de contagio se calculará la "7 day incidence rate", para esto se necesita los casos 
# confirmados en los últimos 7 días. Al usar esta metodología nos permite comparar
# de mejor manera los países pequeños con los grandes, ya que se expresa la tasa en 100,000 habitantes y esto hace que 
# la comparación sea más equitativa.
#Fuente-> Georgia Department of Public Health: https://covid19.gachd.org/daily-case-rate-per-100000/

#Fórmula= (N/P)*100, DONDE N=casos confirmados en los últimos 7 días y P= población
```

```python
seven=sumadead.iloc[:,[454,455,456,457,458,459,460]]
```

```python
seven['casos7dias']=seven.sum(axis=1)
```

```python
numerocases=seven[['casos7dias']]
numerocases.reset_index(inplace=True)
```

```python
numerocases=numerocases.groupby('Country/Region')[['casos7dias']].sum()
```

```python
numerocases
```

```python
numerocases.reset_index(inplace=True)
```

```python
# Para la población se ha buscado una base de datos externa que nos brinda la población mundial por países entre el 2020 y 2021
population=pd.read_csv('UID_ISO_FIPS_LookUp_Table.csv')
```

```python
total=population.groupby('Country_Region')[['Population']].sum()
```

```python
total
```

```python
total.reset_index(inplace=True)
```

```python
total=total.rename(columns={'Country_Region':'Country/Region'})
```

```python
incidencerate=pd.merge(numerocases, total)
```

```python
incidencerate['irate']=100000*incidencerate['casos7dias']/incidencerate['Population']
```

```python
incidencerate
```

```python
incidencerate.irate.sort_values()
```

```python
incidencerate.iloc[77] #Hungary es el país con mayor tasa de contagios diaria
```

### Pregunta 4

```python
# Para este se utilizará la fórmula del método 1 que menciona el artículo del Central Asian Journal of Global Health
# Artículo del Central Asian Journal of Global Health:
# http://cajgh.pitt.edu/ojs/cajgh/article/view/489/315

#El cual calcula el Case Recovery Rate
#Fórmula: CRR= (R/C)*100, DONDE R: total de recuperados de COVID Y C: total de casos COVID
```

```python
recuperados=recovglo.drop(columns=['Province/State', 'Lat', 'Long'])
```

```python
recuperados['totalrecuperados']=recuperados.sum(axis=1)
```

```python
recuperados
```

```python
recover=recuperados.groupby('Country/Region')[['totalrecuperados']].sum()
```

```python
recover.reset_index()
```

```python
recover.reset_index(inplace=True)
```

```python
recovery=pd.merge(recover, positivos)
```

```python
recovery
```

```python
recovery['recovery_rate']=100*recovery['totalrecuperados']/recovery['totalpositivos']
```

```python
recovery
```

```python
recovery.recovery_rate.sort_values()
```

```python
recovery.iloc[184] #Uzbekistan se ha recuperado más rápido.
```
